```python
import pandas as pd

# Replace with your file name if different
df = pd.read_csv("pizza_sales.csv")

df.head()   # Preview data
```

```python
# Total & Average Sales
total_sales = df['total_price'].sum()
average_sales = df['total_price'].mean()
```

```python
# Convert date column
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
# Top Selling Pizza
top_pizza = df.groupby('pizza_name')['total_price'].sum().sort_values(ascending=False)
```

```python
# Sales by Category
category_sales = df.groupby('pizza_category')['total_price'].sum()
```

```python
# Sales by Size
size_sales = df.groupby('pizza_size')['total_price'].sum()
```

```python
# Monthly Sales
monthly_sales = df.groupby(df['order_date'].dt.to_period('M'))['total_price'].sum()
```

```python
# Create Summary DataFrame
summary = pd.DataFrame({
    "Metric": ["Total Sales", "Average Sales"],
    "Value": [total_sales, average_sales]
})
```

```python
# Save everything to Excel
with pd.ExcelWriter("Pizza_Sales_Automation_Report.xlsx") as writer:
    df.to_excel(writer, sheet_name="Raw Data", index=False)
    summary.to_excel(writer, sheet_name="Summary", index=False)
    top_pizza.to_excel(writer, sheet_name="Top Pizza")
    category_sales.to_excel(writer, sheet_name="Category Sales")
    size_sales.to_excel(writer, sheet_name="Size Sales")
    monthly_sales.to_excel(writer, sheet_name="Monthly Sales")

files.download("Pizza_Sales_Automation_Report.xlsx")

print("Automation Report Generated Successfully!")
```
